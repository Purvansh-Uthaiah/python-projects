```python
import pandas as pd
```

```python
data = {
    "subject":['Purvansh','Pavan','Poojith','Kaushali','Hridhan','Parinith'],
    'maths':['88','80','78','96','100','90'],
    'Physics':['94','90','96','88','80','90'],
    'P.E':['92','92','34','70','96','84'],
    'English':['88','70','72','98','68','78'],
    'C.S':['78','76','70','90','80','81'],
    'History':['88','90','88','84','86','88']
}
df = pd.DataFrame(data,index=[1,2,3,4,5,6])
df
```

```python
df.to_csv('Students.csv',index = False)
```

```python
city = ['Madikeri','Madikeri','Mysore',"Bangalore",'Madikeri','Mysore']
df['city'] = city
```

```python
df
```

```python

city_pos = df.columns.get_loc('city')
print(city_pos)
```

```python

subject = ['maths','Physics','P.E','English','C.S','History']

df[subject] = df[subject].apply(pd.to_numeric)

Total = df[subject].sum(axis = 1)
df.insert(city_pos ,"Total" ,Total)
df
```

```python
Percent = df['Total'] / len(subject)
df.insert(city_pos, "Percentage", Percent)
df
```

```python
cols = list(df.columns)
t = cols.index("Total")
p = cols.index("Percentage")
cols[p],cols[t] = cols[t],cols[p]
df = df[cols]
df
```

```python
def get_grade(p):

 if(p >= 90 ):
   return  'A'
 elif(p >= 85):
   return 'B'
 elif(p >=75 ):
   return 'C'
 else:
    return 'D'

df['Grade'] = df['Percentage'].apply(get_grade)
df
```

```python
def pass_fail(row):
    if all(row[s] >= 40 for s in subject):
        return 'PASS'
    else:
        return 'FAIL'

df['Result'] = df.apply(pass_fail,axis = 1)
df
    
```

```python
df['Rank'] = df['Percentage'].rank(ascending = False).astype(int)
df
```

```python
df.to_csv('Results.csv')
```
